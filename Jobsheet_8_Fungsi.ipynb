#Praktikum 04: Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
      print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

#Praktikum 07: Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

#Tugas No 10: Fungsi Decorator

```python
def decorator_logger(fungsi):
  def wrapper():
    fungsi()
    print("Fungsi selesai dijalankan.")
  return wrapper

@decorator_logger
def halo():
  print("Memulai fungsi...")

halo()
```

#Tugas No 11: Studi Kasus Nyata - Program Nilai

```python
def input_nilai():
    nilai_list = []
    while True:
        try:
            nilai = float(input("Masukkan nilai (ketik 'selesai' untuk mengakhiri): "))
            nilai_list.append(nilai)
        except ValueError:
            input_str = input().lower()
            if input_str == 'selesai':
                break
            else:
                print("Input tidak valid. Harap masukkan angka atau 'selesai'.")
    return nilai_list

def hitung_rata(data):
    if not data:
        return 0
    return sum(data) / len(data)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

# Fungsi lambda untuk menghitung nilai akhir
# Bobot: 40% tugas, 60% ujian
hitung_nilai_akhir = lambda tugas, ujian: (0.40 * tugas) + (0.60 * ujian)

print("=== Program Analisis Nilai Mahasiswa ===")

# a. Menerima nilai mahasiswa
print("\n--- Masukkan Nilai Mahasiswa ---")
nilai_mahasiswa = input_nilai()

if nilai_mahasiswa:
    # b. Menghitung rata-rata nilai
    rata_rata = hitung_rata(nilai_mahasiswa)
    print(f"\nRata-rata nilai mahasiswa: {rata_rata:.2f}")

    # c. Menentukan status kelulusan
    status = status_kelulusan(rata_rata)
    print(f"Status kelulusan: {status}")

    # d. Menghitung nilai akhir dengan lambda
    print("\n--- Perhitungan Nilai Akhir (Bobot: 40% Tugas, 60% Ujian) ---")
    try:
        nilai_tugas = float(input("Masukkan nilai tugas: "))
        nilai_ujian = float(input("Masukkan nilai ujian: "))
        nilai_akhir = hitung_nilai_akhir(nilai_tugas, nilai_ujian)
        print(f"Nilai akhir berdasarkan bobot: {nilai_akhir:.2f}")
    except ValueError:
        print("Input nilai tugas atau ujian tidak valid.")
else:
    print("Tidak ada nilai yang dimasukkan.")
```
